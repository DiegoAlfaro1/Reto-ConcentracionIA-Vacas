```python
# Obtener fecha min y max por vaca
datosPorVaca = df.groupby(('ID', 'ID Vaca'))[[('Fecha y hora de inicio', 'fecha')]].agg(['min', 'max']) #[('Fecha y hora de inicio', 'fecha')]

# Para renombrar las columnas para mejor claridad
datosPorVaca.columns = ['PrimeraFecha', 'UltimaFecha']

# Para ver también la cantidad de días entre la primera y última fecha
datosPorVaca['DiasTotales'] = (datosPorVaca['UltimaFecha'] - datosPorVaca['PrimeraFecha']).dt.days

datosPorVaca['DuracionPromedio'] = df.groupby(('ID', 'ID Vaca'))[[('Main', 'Duracion (mm:ss)')]].mean()

# datosPorVaca['DecaemientoDeProduccion'] = 

datosPorVaca.sort_values(by='DiasTotales', ascending=False)
```

```python
# Obtener fecha min y max por vaca
fechas_por_vaca = (
    df.groupby(['ID', 'ID Vaca'])[('Fecha y hora de inicio', 'fecha')]
      .agg(['min', 'max'])
      .reset_index()
)

# Para renombrar las columnas para mejor claridad
fechas_por_vaca.columns = ['Primera_fecha', 'Ultima_fecha']

# Para ver también la cantidad de días entre la primera y última fecha
fechas_por_vaca['Dias_totales'] = (fechas_por_vaca['Ultima_fecha'] - fechas_por_vaca['Primera_fecha']).dt.days

print(fechas_por_vaca)
```
